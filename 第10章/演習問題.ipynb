```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
def hazard_function(y, theta, lambda_):
    return (np.exp(theta) * lambda_ * y ** (lambda_ - 1)) / (
        1 + np.exp(theta) * y**lambda_
    )
```

```python
# Parameters
lambda_values = [1, 5]
theta_values = [-5, -2, 0.5]

# y range
y = np.linspace(0.01, 3, 300)

# Plot
for theta in theta_values:
    plt.figure(figsize=(10, 6))

    for lambda_ in lambda_values:
        h = hazard_function(y, theta, lambda_)
        plt.plot(y, h, label=f"lambda={lambda_}")

    plt.title(f"Hazard Function (theta={theta})")
    plt.xlabel("y")
    plt.ylabel("Hazard")
    plt.grid(True)
    plt.legend()
    plt.savefig(f"plot_{theta}.png")
    plt.show()
```
