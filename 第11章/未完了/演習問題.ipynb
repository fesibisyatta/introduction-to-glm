## pooled analysis

```python
X_pooled = df_result["x"].values
X_pooled = np.stack([np.ones(len(X_pooled)), X_pooled], axis=1)
y_pooled = df_result["y"].values

# 係数推定
beta_pooled = np.linalg.inv(X_pooled.T @ X_pooled) @ X_pooled.T @ y_pooled

# 不偏分散の推定
sigma_hat_sq = ((y_pooled - X_pooled @ beta_pooled) ** 2).sum() / (
    df_result.shape[0] - 2
)

# 標準誤差の推定
std_error_pooled = np.sqrt(np.diag(np.linalg.inv(X_pooled.T @ X_pooled)) * sigma_hat_sq)

# データフレームにまとめる
df_coef_pooled = pd.DataFrame(
    np.concatenate([beta_pooled, std_error_pooled]).reshape(4, 1)
).T
df_coef_pooled.columns = [
    "intercept",
    "slope",
    "std_error(intercept)",
    "std_error(coef)",
]
df_coef_pooled.index = ["pooled analyasis"]
```

## データ縮約法による探索的解析

### dog と反応変数が独立の場合

```python
list_coef = []
for dog, group in df_result.groupby("dog"):
    # デザイン行列の作成
    X_reduction = group["x"].values
    X_reduction = np.stack([np.ones(len(X_reduction)), X_reduction], axis=1)

    # 目的変数
    y_reduction = group["y"].values

    # 係数推定
    beta_reduction = (
        np.linalg.inv(X_reduction.T @ X_reduction) @ X_reduction.T @ y_reduction
    )
    beta_reduction = beta_reduction.tolist()

    # 不偏分散
    sigma_sq_hat = ((y_reduction - X_reduction @ beta_reduction) ** 2).sum() / (
        len(X_reduction) - 2
    )

    # 標準誤差
    std_error_reduction = np.sqrt(
        np.diag(np.linalg.inv(X_reduction.T @ X_reduction)) * sigma_sq_hat
    ).tolist()

    list_coef.append([dog] + beta_reduction + std_error_reduction)

df_coef_ignore_dog = pd.DataFrame(
    list_coef,
    columns=["dog", "intercept", "slope", "std_error(intercept)", "std_error(coef)"],
)
```

### condition と反応変数が独立の場合

```python
list_coef = []
for cond, group in df_result.groupby("condition"):
    # デザイン行列の作成
    X_reduction = group["x"].values
    X_reduction = np.stack([np.ones(len(X_reduction)), X_reduction], axis=1)

    # 目的変数
    y_reduction = group["y"].values

    # 係数推定
    beta_reduction = (
        np.linalg.inv(X_reduction.T @ X_reduction) @ X_reduction.T @ y_reduction
    )
    beta_reduction = beta_reduction.tolist()

    # 不偏分散
    sigma_sq_hat = ((y_reduction - X_reduction @ beta_reduction) ** 2).sum() / (
        len(X_reduction) - 2
    )

    # 標準誤差
    std_error_reduction = np.sqrt(
        np.diag(np.linalg.inv(X_reduction.T @ X_reduction)) * sigma_sq_hat
    ).tolist()
    list_coef.append([cond] + beta_reduction + std_error_reduction)

df_coef_ignore_cond = pd.DataFrame(
    list_coef,
    columns=[
        "condition",
        "intercept",
        "slope",
        "std_error(intercept)",
        "std_error(coef)",
    ],
)
```

## GEEモデルによる推定 

### dogがsubjectと仮定
https://www.maths.usyd.edu.au/u/jchan/GLM/Liang&Zeger1986GEE.pdf の(8)式による


```python
df_result.groupby("dog").count()
```

```python
def irls_for_gee(df_result, is_sandwitch, unbaised_variance):
    df_result = df_result.copy()

    # 初期値
    beta = np.array([1.0, 1.0])

    # 特徴量
    p = 2

    # 各被験者は8回の繰り返しがあり、dogの種類は5種類
    # 作業相関行列およびAの初期値は単位行列とする
    R = [np.eye(8) for _ in range(df_result["dog"].nunique())]
    A = [np.eye(8) for _ in range(df_result["dog"].nunique())]
    psi = 1

    # 10回IRLSで更新
    for iter_ in range(20):
        D = []
        V_inv = []
        S = []
        pearson_resid = []
        y = []

        for i, (dog, group) in enumerate(df_result.groupby("dog")):
            # muの勾配
            Di = np.stack([np.ones(len(group)), group["x"].values], axis=1)
            D.append(Di)

            # yiの推定
            # 実際には link function の逆関数を噛ませる必要があるが、
            # リンク関数として恒等関数を選択したのでそのままの値
            mu_i_hat = (Di * beta).sum(axis=1)

            # 第i groupにおける y の値
            yi = group["y"].values
            y.append(yi)

            # 残差
            resid = yi - mu_i_hat
            S.append(resid)

            # 正規分布の分散
            # 不偏分散を毎回iterごとに更新したものにするのも良いかも
            if unbaised_variance:
                variance = (resid**2).sum() / (8 - p)
            else:
                variance = 1

            # ピアソン残差
            ri = resid / np.sqrt(variance)
            pearson_resid.append(ri)

            # 分散の推定
            V_inv_i = np.sqrt(A[i]) @ R[i] @ np.sqrt(A[i]) * psi
            V_inv.append(np.linalg.inv(V_inv_i))

        if is_sandwitch:
            # 情報行列の計算
            J = np.zeros((2, 2))
            for i in range(len(D)):
                J += D[i].T @ V_inv[i] @ D[i]

            # Cの計算
            C = np.zeros((2, 2))
            for i in range(len(D)):
                C += D[i].T @ V_inv[i] @ np.outer(S[i], S[i]) @ V_inv[i] @ D[i]

            # サンドイッチ推定量
            std_error = np.linalg.inv(J) @ C @ np.linalg.inv(J)

        else:
            std_error = np.zeros((2, 2))
            for i in range(len(D)):
                std_error += D[i].T @ V_inv[i] @ D[i]
            std_error = np.linalg.inv(std_error)

        # subjectの数だけ足し合わせる
        mat_1 = np.zeros((2, 2))
        mat_2 = np.zeros((2, 1))
        for i in range(len(D)):
            mat_1 += D[i].T @ V_inv[i] @ D[i]
            mat_2 += D[i].T @ V_inv[i] @ S[i].reshape(8, 1)

        # IRLSによる更新
        beta += (np.linalg.inv(mat_1) @ mat_2).reshape(-1)

        # beta更新後にR, A, psiを更新する
        # mu_hat_i = (Di*beta).sum(axis = 1)
        psi = sum((pearson_resid[i] ** 2).sum() for i in range(len(pearson_resid))) / (
            len(df_result) - p
        )

        for i in range(len(pearson_resid)):
            Ri = np.outer(pearson_resid[i], pearson_resid[i]) / (len(df_result) - p)
            R[i] = Ri
            A[i] = np.diag(np.diag(Ri))

    return beta, np.diag(std_error)
```

```python
is_sandwitch = True
unbaised_variance = True
beta, std_error = irls_for_gee(df_result, is_sandwitch, unbaised_variance)
```
