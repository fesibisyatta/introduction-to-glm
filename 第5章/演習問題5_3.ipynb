```python
import numpy as np
import pandas as pd
```

```python
rng = np.random.default_rng(seed=123)

# [0, 1]区間における乱数生成
length = 100
uniform_random_number = rng.random(length)

# Pareto分布に従う乱数
theta = 2
Y = (1 / uniform_random_number)**(1/theta)
Y
```

```python
# thetaの最尤推定値
theta_est = 1/np.mean(np.log(Y))
theta_est
```

```python
# 20回繰り返し信頼区間を求める
n_iters = 20
theta = 2

list_theta_est = []
for iter_ in range(n_iters):
    uniform_random_number = rng.random(length)
    
    # Pareto分布に従う乱数
    
    Y = (1 / uniform_random_number)**(1/theta)
    
    # thetaの最尤推定値
    theta_est = 1/np.mean(np.log(Y))
    list_theta_est.append(theta_est)
list_theta_est = np.array(list_theta_est)

df_est = pd.DataFrame(
    {
        "true":theta,
        "estimate":list_theta_est,
        "lower_CI95%":(list_theta_est / (1 + 1.96/np.sqrt(length))),
        "upper_CI95%":(list_theta_est / (1 - 1.96/np.sqrt(length)))
    }
).round(2)
```

```python
# theta_estの平均値と真のthetaとの比
df_est.mean().round(3)
```
