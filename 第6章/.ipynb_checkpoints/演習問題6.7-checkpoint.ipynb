```python
# (a)
df = pd.DataFrame(
    {
        "時期": ["1936-39", "1946-49", "1956-59", "1966-69", "1976-79", "1986-89"],
        "精糖の消費量": [32, 31.2, 27, 21, 14.9, 8.8],
        "加工食品に含まれる砂糖の消費量": [16.3, 23.1, 23.6, 27.7, 34.6, 33.9],
    }
)

plt.title("精糖及び加工食品に含まれる砂糖の消費量")
plt.scatter(df["時期"], df["精糖の消費量"], label="精糖")
plt.scatter(df["時期"], df["加工食品に含まれる砂糖の消費量"], label="砂糖")
plt.legend()
plt.savefig("test1.png")

plt.show()
```

```python
df_concat = pd.concat([df, pd.get_dummies(df["時期"]).astype(int)], axis=1)
df_concat = sm.add_constant(df_concat)

df_concat["期間"] = list(range(1, len(df_concat) + 1))
```

```python
X = df_concat[["const", "加工食品に含まれる砂糖の消費量"]].values
```

```python
from scipy.stats import t

alpha = 0.05
n, p = X.shape
t_crit = t.ppf(1 - alpha / 2, df=n - p)

X = np.stack([np.ones(6), df_concat["期間"].values], axis=1)
y = df_concat["加工食品に含まれる砂糖の消費量"].values

# 係数
beta = np.linalg.inv(X.T @ X) @ X.T @ y

# 標準誤差の計算
sigma_hat_sq = ((y - X @ beta).T @ (y - X @ beta)) / (n - p)
std_error = np.diag(np.sqrt(np.linalg.inv(X.T @ X) * sigma_hat_sq))

# 表に出力
table = pd.DataFrame(beta, index=["const", "係数"], columns=["推定値"])
table["標準誤差"] = std_error
table["95%信頼区間下限"] = beta - t_crit * table["標準誤差"]
table["95%信頼区間上限"] = beta + t_crit * table["標準誤差"]
table
```

## (b)

```python
df_616["定数項"] = 1
df_616["log1p_K"] = np.log1p(df_616["K"])
df_616["log1p_収穫量"] = np.log1p(df_616["収穫量"])
df_616["K_squared"] = df_616["K"] ** 2

# 通常の線形回帰
X_normal = df_616[["定数項", "K"]].values
y_normal = df_616["収穫量"].values

# リン総量の対数変換
X_log_lin = df_616[["定数項", "log1p_K"]].values
y_log_lin = df_616["収穫量"].values

# リン総量と収穫量の対数変換
X_log_lin_and_yield = df_616[["定数項", "log1p_K"]].values
y_log_lin_and_yield = df_616["log1p_収穫量"].values

# リン総量の二乗項を追加
X_quad = df_616[["定数項", "K", "K_squared"]].values
y_quad = df_616["収穫量"].values

models = {
    "線形回帰": (X_normal, y_normal),
    "線形回帰(リン総量対数変換あり)": (X_log_lin, y_log_lin),
    "線形回帰(リン総量, 収穫量対数変換あり)": (X_log_lin_and_yield, y_log_lin_and_yield),
    "2次多項式モデル": (X_quad, y_quad),
}

dict_coef={}
dict_std_residuals={}
for model_name, (X, y) in models.items():
    # 係数
    beta = np.linalg.inv(X.T @ X) @ X.T @ y
    dict_coef[model_name]=beta

    # 予測値
    y_pred = X @ beta

    # 残差
    residuals = y - y_pred

    # 決定係数
    ss_total = np.sum((y - np.mean(y)) ** 2)
    ss_residual = np.sum(residuals ** 2)
    r_squared = 1 - (ss_residual / ss_total)

    # 分散の最尤推定量
    n, p = X.shape
    sigma_sq_mle = np.sum(residuals ** 2) / (X.shape[0] - X.shape[1])

    # ハット行列
    H = np.eye(len(X)) - X @ np.linalg.inv(X.T @ X) @ X.T

    # 標準化残差
    r_i = residuals / np.sqrt((1 - np.diag(H)) * sigma_sq_mle)
    dict_std_residuals[model_name]=r_i

    # 対数尤度
    n = len(y)
    sigma_sq = np.sum(residuals ** 2) / n
    log_likelihood = -0.5 * n * (np.log(2 * np.pi) + np.log(sigma_sq) + 1)

    # 対数尤度比統計量の計算
    

    print(f"{model_name} の対数尤度: {log_likelihood:.4f}")

    print(f"{model_name} の決定係数: {r_squared:.4f}")
```

```python

for model in dict_coef.keys():
    print(f"{model} の係数: {dict_coef[model]}")
```

```python
data = {
    'Carbohydrate': [33, 40, 37, 27, 30, 43, 34, 48, 30, 38, 50, 51, 30, 36, 41, 42, 46, 24, 35, 37],
    'Age': [33, 47, 49, 35, 46, 52, 62, 23, 32, 42, 31, 61, 63, 40, 50, 64, 56, 61, 48, 28],
    'Weight': [100, 92, 135, 144, 140, 101, 95, 101, 98, 105, 108, 85, 130, 127, 109, 107, 117, 100, 118, 102],
    'Protein': [14, 15, 18, 12, 15, 15, 14, 17, 15, 14, 17, 19, 19, 20, 15, 16, 18, 13, 18, 14]
}
df = pd.DataFrame(data)
```

```python
df
```

## (a)

```python
for v in ["Age", "Weight", "Protein"]:

    plt.figure(figsize=(7, 4))
    plt.scatter(df[v], df["Carbohydrate"], color='blue', label='Data points')
    
    correlation = df[v].corr(df["Carbohydrate"])
    slope, intercept = np.polyfit(df[v], df["Carbohydrate"], 1)
    x_range = np.array([df[v].min(), df[v].max()])
    y_pred = slope * x_range + intercept
    
    plt.plot(x_range, y_pred, color='red', linestyle='--', label=f'Regression Line')
    
    plt.title(f"{v} vs Carbohydrate (corr: {correlation:.2f})")
    plt.xlabel(v)
    plt.ylabel("Carbohydrate")
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend()
    plt.savefig(v + ".png")
    plt.show()
```

## (b)

```python
#本文p112に記載されている係数を流用
b = [36.9601, -0.1137, -0.2280, 1.9577]

# 当てはめ値
df["pred_Carbohydrate"] = b[0] + b[1] * df["Age"] + b[2] * df["Weight"] + b[3] * df["Protein"]

# 残差
df["resid"] = df["Carbohydrate"] - df["pred_Carbohydrate"]

# 不偏分散
sigma_sq_hat = (df["resid"]**2).sum() / (df.shape[0] - 4)

# ハット行列
X = np.column_stack([np.ones(len(df)).reshape(-1, 1), df[["Age", "Weight", "Protein"]].values])
H = np.eye(len(X)) - X @ np.linalg.inv(X.T@X) @ X.T

# 標準化残差
df["std_resid"] = df["resid"].values/np.sqrt(sigma_sq_hat * (1 - np.diag(H)))
```

```python
for v in ["Age", "Weight", "Protein", "pred_Carbohydrate"]:

    plt.figure(figsize=(7, 4))
    plt.scatter(df[v], df["std_resid"], color='blue', label='Data points')
    
    correlation = df[v].corr(df["std_resid"])
    slope, intercept = np.polyfit(df[v], df["std_resid"], 1)
    x_range = np.array([df[v].min(), df[v].max()])
    y_pred = slope * x_range + intercept
    
    plt.plot(x_range, y_pred, color='red', linestyle='--', label=f'Regression Line')
    
    plt.title(f"{v} vs std_resid (corr: {correlation:.2f})")
    plt.xlabel(v)
    plt.ylabel("std_resid")
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend()
    plt.savefig(f"std_residual_vs_{v}.png")
    plt.show()
```
