```python
import pandas as pd

data = [
    # Left block
    [5.94, 52, 20.7],
    [4.71, 46, 21.3],
    [5.86, 51, 25.4],
    [6.52, 44, 22.7],
    [6.80, 70, 23.9],
    [5.23, 33, 24.3],
    [4.97, 21, 22.2],
    [8.78, 63, 26.2],
    [5.13, 56, 23.3],
    [6.74, 54, 29.2],
    [5.95, 44, 22.7],
    [5.83, 71, 21.9],
    [5.74, 39, 22.4],
    [4.92, 58, 20.2],
    [6.69, 58, 24.4],

    # Right block
    [6.48, 65, 26.3],
    [8.83, 76, 22.7],
    [5.10, 47, 21.5],
    [5.81, 43, 20.7],
    [4.65, 30, 18.9],
    [6.82, 58, 23.9],
    [6.28, 78, 24.3],
    [5.15, 49, 23.8],
    [2.92, 36, 19.6],
    [9.27, 67, 24.3],
    [5.57, 42, 22.0],
    [4.92, 29, 22.5],
    [6.72, 33, 24.1],
    [5.57, 42, 22.7],
    [6.25, 66, 27.3],
]

df = pd.DataFrame(data, columns=["CHOL", "Age", "BMI"])
```

```python
# 目的変数
y = df["CHOL"].values

# 説明変数( Age + BMI )
X1 = df[["Age", "BMI"]].values
X1 = np.column_stack([np.ones(len(y)), X1]) # 切片を追加

# 回帰係数
beta1 = np.linalg.inv(X1.T @ X1) @ X1.T @ y

# 逸脱度
D1 = y.T @ y - beta1 @ X1.T @y

# 自由度
deg_o_f_1 = X1.shape[0] - X1.shape[1] 

# 説明変数(Age)
X2 = df[["Age"]].values
X2 = np.column_stack([np.ones(len(y)), X2]) # 切片を追加

# 回帰係数
beta2 = np.linalg.inv(X2.T @ X2) @ X2.T @ y

# 逸脱度
D2 = y.T @ y - beta2 @ X2.T @y

# 自由度
deg_o_f_2 = X2.shape[0] - X2.shape[1]

# F値
F = ((D2 - D1)/(deg_o_f_2 - deg_o_f_1) )/ (D1/deg_o_f_1)
```

| Age + BMI モデル | 26.57062624947889        | 27          |
| Age モデル       | 31.636154102712453       | 28          |
