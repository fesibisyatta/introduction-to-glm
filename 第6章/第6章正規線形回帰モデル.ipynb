```python
sigma_hat_sq = (y.T @ y - y.T @ X @ b - b.T @ X.T @ y + b.T @ X.T @ X @ b) / (
    len(df) - len(X.columns)
)
sigma_hat_sq
```

```python
X_0 = X.drop(["年齢"], axis=1)
b_0 = np.linalg.inv(X_0.T @ X_0) @ (X_0.T @ y)

p = len(X.columns)
q = len(X_0.columns)

print("年齢なしのモデルによる回帰係数：", b_0)

df_tmp = pd.DataFrame(
    {
        "変動要因": ["モデル(6.7)", "モデル(6.6)による改善", "残差"],
        "自由度": [q, p - q, len(df) - p],
        "平方和": [
            round(b_0.T @ X_0.T @ y, 3),
            round(b.T @ X.T @ y - b_0.T @ X_0.T @ y, 3),
            round(((y - X @ b) ** 2).sum(), 3),
        ],
        "平均平方": [
            np.nan,
            round((b.T @ X.T @ y - b_0.T @ X_0.T @ y) / (p - q), 2),
            round(((y - X @ b) ** 2).sum() / (len(df) - p), 2),
        ],
    }
)
pd.concat(
    [
        df_tmp,
        pd.DataFrame(
            {
                "変動要因": ["計"],
                "自由度": [len(df)],
                "平方和": [
                    round(b_0.T @ X_0.T @ y, 3)
                    + round(b.T @ X.T @ y - b_0.T @ X_0.T @ y, 3)
                    + round(((y - X @ b) ** 2).sum(), 3)
                ],
                "平均平方": np.nan,
            }
        ),
    ],
    axis=0,
).reset_index(drop=True)
```
